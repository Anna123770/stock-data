```python
#daily stock price trend chart
import matplotlib.pyplot as plt

plt.style.use('seaborn')
fig,ax=plt.subplots(3)
df_AAPL.Close.plot(ax=ax[0],color='red',legend=True)
df_GOOG.Close.plot(ax=ax[1],color='green',legend=True)
df_META.Close.plot(ax=ax[2],color='blue',legend=True)
plt.show()
```

```python
#put three charts togethor
df_AAPL.Close.plot(legend=True)
df_GOOG.Close.plot(legend=True)
df_META.Close.plot(legend=True)
plt.show()
```

```python
#daily turnover trend chart
fig,ax=plt.subplots(3)
df_AAPL.Volume.plot(ax=ax[0],color='red',legend=True)
df_GOOG.Volume.plot(ax=ax[1],color='green',legend=True)
df_META.Volume.plot(ax=ax[2],color='blue',legend=True)
plt.show()
```

```python
#company info
company={'name':['AAPL','GOOG','META'],'industry':['Consumer Electronics','Internet Content & Information','Internet Content & Information'],'total_revenue':[365817,257637,117929],'net_income':[94680,76033,39370]}
df=pd.DataFrame(company)
print(df)
```

```python
# market value & daily turnover
from turtle import color


fig,ax=plt.subplots(1,2)
df.total_revenue.plot.bar(ax=ax[0],color='black',legend=True).set_xticks([0,1,2],['AAPL','GOOG','META'],rotation=30)
df.net_income.plot.bar(ax=ax[0],color='purple',legend=True).set_xticks([0,1,2],['AAPL','GOOG','META'],rotation=30)
df_AAPL.Volume.plot(ax=ax[1],color='red',legend=True)
df_GOOG.Volume.plot(ax=ax[1],color='green',legend=True)
df_META.Volume.plot(ax=ax[1],color='blue',legend=True)
plt.show()
```

```python
# Risk & Return forecasting
df_AAPL['daily_ret']=df_AAPL.Adj_Close.pct_change()
df_GOOG['daily_ret']=df_GOOG.Adj_Close.pct_change()
df_META['daily_ret']=df_META.Adj_Close.pct_change()
```

```python
df_AAPL.daily_return.plot(color='red',legend=True)
df_GOOG.daily_return.plot(color='green',legend=True)
df_META.daily_return.plot(color='blue',legend=True)
plt.show()
```

```python
print(f'At 95% confidence interval,the loss isn\'t exceed.\nAAPL: {abs(df_AAPL.daily_return.quantile(0.05)*100):.3f}%\nGOOG: {abs(df_GOOG.daily_return.quantile(0.05)*100):.3f}%\nMETA: {abs(df_META.daily_return.quantile(0.05)*100):.3f}%')
```
